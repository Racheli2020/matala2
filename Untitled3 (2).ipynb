```python
#working q2

from datetime import datetime, timedelta

data2.loc[:, 'event_time'] = pd.to_datetime(data2['event_time'])

data2 = data2.sort_values(by=['user_session', 'event_time'])

def calculate_funnel_number(group):
    event1 = group['event_time'].iloc[0]
    event2 = group['event_time'].shift(1).iloc[0]
    time_diff = event1 - event2
    if time_diff.days <= 5:
        return 1
    else:
        time_diffs = (group['event_time'] - group['event_time'].shift(1)).dt.days.fillna(0)
        time_diffs.loc[time_diffs > 5] = 1
        return time_diffs.cumsum() + 1

data2['funnel_number'] = data2.groupby('user_session').apply(calculate_funnel_number).reset_index(drop=True)

data2


```
